```python
import scipy as sp
import scipy.stats
import matplotlib.pylab as plt
```

```python
# 정규분포를 이루는 확률변수 생성
# 모수 지정
# loc=분포의 기댓값
# scale=분포의 폭 (표준 편차)
rv=sp.stats.norm(loc=1,scale=2)
```

```python
# 확률분포 매서드
# pdf : 확률밀도 함수
#  함수기 때문에 값을 넣어야 결과 확인이 가능
xx=np.linspace(-8,8,100)
pdf=rv.pdf(xx)
plt.plot(xx,pdf)
plt.title("확률밀도 함수")
plt.xlabel("$x$")
plt.ylabel("$p(x)$")
plt.show()
```

```python
# 연습문제 8.1.1
# 기댓값이0  표준편차 0.1인 정규분포 만들고 확률 밀도함수 그려라
rv=sp.stats.norm(loc=0,scale=0.1)
xx=np.linspace(-1,1,100)
pdf=rv.pdf(xx)
plt.plot(xx,pdf)
plt.title("확률밀도 함수")
plt.xlabel("$x$")
plt.ylabel("$p(x)$")
plt.show()
```

```python
# 이 확률밀도함수의 최댓값은 얼마인가
# x=0에서 최댓값을 가진다.
# 이값은 확률은 아님
rv.pdf(0)
```

```python
# 누적분포 함수
# 여기는 f(x)값이 진짜 확률값
rv=sp.stats.norm(loc=0,scale=1)
xx = np.linspace(-8, 8, 100)
cdf = rv.cdf(xx)
plt.plot(xx, cdf)
plt.title("누적분포함수 ")
plt.xlabel("$x$")
plt.ylabel("$F(x)$")
plt.show()
```

```python
# 해당 확률변수에서 무작위 표본을 생성 rvs
# random_state -> seed
rv.rvs(size=(3,5),random_state=0)
```

```python
# 변환확률 변수의 시뮬레이션
#  0과 1 사이의 균일분포를 가지는 확률변수에서 두 표본값을 생성하여 
# 이 두 값을 합하면 결과는 어떤 분포를 가질까?
rv1 = sp.stats.uniform()
rv2 = sp.stats.uniform()

np.random.seed(0)
N = 50000
x_1 = rv1.rvs(N)
x_2 = rv2.rvs(N)
x_3 = x_1 + x_2

plt.figure(figsize=(12, 5))

plt.subplot(131)
sns.distplot(x_1, kde=False)
plt.title("균일분포")
plt.xlabel("표본값")
plt.xlim(-0.2, 2.2)

plt.subplot(132)
sns.distplot(x_2, kde=False)
plt.title("균일분포")
plt.xlabel("표본값")
plt.xlim(-0.2, 2.2)

plt.subplot(133)
sns.distplot(x_3, kde=False)
plt.title("균일분포 표본의 합의 분포")
plt.xlabel("표본값")
plt.xlim(-0.2, 2.2)

plt.show()
```

```python
# 연습 문제 8.1.3¶
# 균일분포 확률분포에서 두 개가 아닌 10개 표본값을 생성하여 그 값의 합을 구하면 
# 어떤 모양의 분포를 이루는지 시뮬레이션 기능을 사용하여 구하라. 
# 이 때 시뮬레이션은 1000번 반복한다.

# 10개 생성을 10000번 반복
# 10개씩 10000행...인듯..
x=sp.stats.uniform().rvs((10,10000))
# 10개 표본의 합이 10000개
s=x.sum(axis=0)
sns.distplot(s)
plt.show()
```

### 8.2 베르누이분포와 이항분포

```python
mu = 0.6
rv = sp.stats.bernoulli(mu)
type(rv)
```

```python
# 이산이라 pmf
# m가 0.6이라 1일때 0.6의 확률을 가짐
xx = [0, 1]
plt.bar(xx, rv.pmf(xx))
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1], ["x=0", "x=1"])
plt.xlabel("표본값")
plt.ylabel("P(x)")
plt.title("베르누이 확률변수의 확률질량함수")
plt.show()
```

```python
# 동전을 100번던진것과 같음
x = rv.rvs(100, random_state=0)
x
```

```python
# 100번의 0.6인 60번정도의 1이 나와야함
sns.countplot(x)
plt.title("베르누이 확률변수의 시뮬레이션 결과")
plt.xlabel("표본값")
plt.show()
```

```python
y = np.bincount(x, minlength=2) / float(len(x))

df = pd.DataFrame({"이론": rv.pmf(xx), "시뮬레이션": y})
df.index = [0, 1]
df
```

```python
df2 = df.stack().reset_index()
df2.columns = ["표본값", "유형", "비율"]
df2
```

```python
sns.barplot(x="표본값", y="비율", hue="유형", data=df2)
plt.title("베르누이분포의 이론적 분포와 시뮬레이션 분포")
plt.show()
```

```python
# 표본평균은 평균의 근삿값이 나옴
np.mean(x)
```

### 이항분포

```python
N = 10
mu = 0.6
rv = sp.stats.binom(N, mu)
```

```python
xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align="center")
plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("이항분포의 확률질량함수")
plt.show()
```

```python
# mu가 0.9라면
# 이항분포는 베르누이 시행을 N번 반복 했을때 N번중 성공한 횟수!
mu=0.9
rv = sp.stats.binom(N, mu)
xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align="center")
plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("이항분포의 확률질량함수")
plt.show()
```

```python
np.random.seed(0)
x = rv.rvs(100)
x
```

```python
sns.countplot(x)
plt.title("이항분포의 시뮬레이션 결과")
plt.xlabel("표본값")
plt.show()
```
