```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
N = 10  # 앞의 10송이만 선택
X = iris.data[:N, :2]
X
```

```python
from sklearn.decomposition import PCA
```

```python
pca1 = PCA(n_components=1)
```

```python
X_low = pca1.fit_transform(X) #10개짜리 1차원 데이터 생성 (크기를 나타내는변수)
# 원점이 기준이 아닌 평균점이 기준으로 되어있어 양수 음수가 다 나타나는...
X_low
```

```python
X2 = pca1.inverse_transform(X_low) 
```

```python
# 축소시킨 차원상의 좌표를 보여주는듯,,
X2
```

```python
# 중심 위치 , 데이터의 평균값
pca1.mean_
```

```python
# 주성분 벡터 , 근사 데이터를 만드는 단위 기저벡터
pca1.components_
```

```python
# 첫번째 오른쪽 특이벡터 구해보면 일치하는 것을 알 수 있음
# 고유벡터의 부호(방향)은 반대가 될 수 도 있다.
X0 = X - X.mean(axis=0)
U, S, VT = np.linalg.svd(X0)
VT
```

```python
# 분산행렬의 고유벡터와도 같음
XCOV = X0.T @ X0
W, V = np.linalg.eig(XCOV)
```

```python
W,V
```

```python
# 8번째 꽃의 꽃받침 길이와 꽃받침 폭
X[7, :]
```

```python
# PCA로 구한 주성분의 값 즉, 꽃의 크기
# 2차원 값이 1차원인 크기로 대체가 됨
X_low[7]
```

```python
# 위 값은 이렇게 구할 수도 있음
pca1.components_ @ (X[7, :] - pca1.mean_)
```

### 이미지 PCA

```python
from sklearn.datasets import fetch_olivetti_faces

faces_all = fetch_olivetti_faces()
K = 20  # 20번 인물의 사진만 선택
faces = faces_all.images[faces_all.target == K]
```

```python
import matplotlib.pylab as plt
```

```python
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(faces[k], cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.suptitle("올리베티 얼굴 사진")
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca3 = PCA(n_components=2)
X3 = faces_all.data[faces_all.target == K]
W3 = pca3.fit_transform(X3)
X32 = pca3.inverse_transform(W3)
```

```python
face_mean = pca3.mean_.reshape(64, 64)
face_p1 = pca3.components_[0].reshape(64, 64)
face_p2 = pca3.components_[1].reshape(64, 64)
```

```python
plt.subplot(131)
plt.imshow(face_mean, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("평균 얼굴")
plt.subplot(132)
plt.imshow(face_p1, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("주성분 1")
plt.subplot(133)
plt.imshow(face_p2, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("주성분 2")
plt.show()
```

```python
# 바라본 방향의 차이
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5 * (k - 5) if k < 5 else 1.5 * (k - 4)
        ax.imshow(face_mean + w * face_p1, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title("주성분1의 비중={}".format(w))
plt.suptitle("평균 얼굴에 주성분1을 더한 사진")
plt.tight_layout()
plt.show()
```

```python
# 웃는것과 웃지않는거의 차이를 나타냄
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5 * (k - 5) if k < 5 else 1.5 * (k - 4)
        ax.imshow(face_mean + w * face_p2, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title("주성분2의 비중={:.1f}".format(w))
plt.suptitle("평균 얼굴에 주성분2를 더한 사진")
plt.tight_layout()
plt.show()
```

### 함수의 그래프

```python
import matplotlib.pylab as plt
```

```python
def f(x):
    return x**3 - 3 * x**2 + x
```

```python
x = np.linspace(-1, 3, 9)
x
```

```python
y = f(x)
y
```

```python
plt.plot(x, y, 'ro-')
plt.xlim(-2, 4)
plt.title("함수 $f(x) = x^3 - 3x^2 + x$의 그래프")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.show()
```

```python
# 파이썬에는 부드러운 곡선을 바로 그릴수 있는 기능이 없음
# 점을 조밀하게 찍어서 그릴수 있음
x = np.linspace(-1, 3, 400)
y = f(x)
plt.plot(x, y)

plt.xlim(-2, 4)
plt.title("함수 $f(x) = x^3 - 3x^2 + x$의 그래프")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))

plt.show()
```

```python
x = np.linspace(-1, 1, 1000) 
```

```python
def heaviside_step(x):     
    if isinstance(x, np.ndarray):         
        return np.where(x >= 0, 1, 0)     
    else:         
        return 1.0 if x == 0 else 0.0
    
# isinstance(인스턴스, 클래스/데이터타입) 타입이나 자료형 일치여부 확인해주는 함수
# np.where(조건, true일때, False일때) for문보다 빠르게 작업할 수 있음 
# 추가로,, np.argmin np.argmax min,max값의 위치 찾기
```

```python
sign=np.sign(x)
heavi=heaviside_step(x)
```

```python
plt.figure(figsize=(7, 4)) 
plt.subplot(121) 
plt.plot(x, sign, "r-") 
plt.title("부호 함수") 
plt.ylim(-1.5, 1.5) 
```

```python
plt.subplot(122)
plt.plot(x, heavi, "k-") 
plt.title("헤비사이드 스텝 함수") 
plt.ylim(-1.5, 1.5) 
plt.tight_layout() 
plt.show()
```

### 수치미분

```python
from scipy.misc import derivative
```

```python
def f(x):
    return x**3 - 3 * x**2 + x

x = np.linspace(-1, 3, 400)
y = f(x)
```

```python
# 3차 다항식의 기울기를 x=0 , x=1지점에서 수치미분으로 구하는 코드(dx는 1/100만)
print(derivative(f, 0, dx=1e-6))
print(derivative(f, 1, dx=1e-6))
```
